## Count of missing features per column. 

```python
df = pd.read_csv("../Data/derived_features_validated.csv")
print("Original DF Length",len(df),"\n")

count_nan = len(df) - df.count()
print("  Number of NA Values Per Column")
print("-----------------------------------")
print(count_nan)

df.dropna(inplace=True)
print("\n Complete Cases length",len(df))
```

```python
cleaned = df.dropna()
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(12,12))
sns.heatmap(df.corr())
ax.set_title("Correlation Matrix", fontsize=16)
```

```python
fig, ((ax1, ax2),(ax3,ax4)) = plt.subplots(nrows=2, ncols=2,figsize=(12,8))
ax1.hist(df['age'].values, bins=9011)
ax1.set_title("Age")


df['gender'].value_counts(sort=True).plot.barh(ax=ax2)
ax2.set_title("Gender")


df['education'].value_counts().plot.barh(ax=ax3)
ax3.set_title("Education")

df['working'].value_counts(sort=True).plot.barh(ax=ax4)
ax4.set_title("Working Status")

plt.tight_layout()
```
